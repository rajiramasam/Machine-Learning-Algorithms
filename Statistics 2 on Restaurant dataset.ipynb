```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('/content/drive/MyDrive/Datasets_ML/Data.csv')
d.head()
```

on which date , the average sales is highest

```python
df=pd.DataFrame(d)
```

```python
x1=df.groupby('Date')['Final Total'].mean().sort_values(ascending=False).head(1)
x1
```

How many total bills are there?

```python
x2=df.groupby('Invoice No.')['Tax'].count()
x2.count()
```

Most sold item of September month

```python
d['Date'].unique()
```

```python
x3=d[(d['Date']>='2023-09-01') & (d['Date']<='2023-09-30')]
x3
```

Which 3 category item is sold less in month of july

```python
j=d[(d['Date']>='2023-07-01') & (d['Date']<='2023-07-31')]
cj=j.groupby('Category')['Qty.'].sum().sort_values(ascending=True).head(3)
cj
```

```python
plt.bar(cj.index,cj.values)
```

Pie chart for sales by category

```python
sc=df.groupby('Category')['Final Total'].sum()
plt.pie(sc,labels=sc.index)
```
